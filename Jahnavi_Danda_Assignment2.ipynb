```python
import numpy as np
#NumPy array of size 15 having Random Integers in the range 1-20
x = np.random.randint(1,21,15)
print("Initial Array\n",x)
#Reshape array to 3 by 5
y= x.reshape(3,5)
#array after reshaping
print('Array after reshaping\n',y)
#Print array shape
print("Array shape is:",y.shape)
#find max element in the array in each row
maxNum=np.amax(y,axis=1)
#Replace the max in each row by 0
print("Array after replacing max ele in row with 0 \n",np.where(np.isin(y,maxNum),0,y))
```

```python
import pandas as pd
#Read ‘data.csv’.
df=pd.read_csv("data.csv")
#Show the basic statistical description about the data.
print("Statistical Description about the data\n",df.describe())
#Check if the data has null values and Replace the null values with the mean
df=df.fillna(df.mean())
#Select at least two columns and aggregate the data using: min, max, count, mean
print("aggregation of the data for the columns duration and pulse using: min, max, count, mean\n",
      df[['Duration','Pulse']].agg(['min','max','mean','count']))
#Filter the dataframe to select the rows with calories values between 500 and 1000. 
print("rows with calories values between 500 and 1000\n",df[(df.Calories>500) & (df.Calories<1000)])
#Filter the dataframe to select the rows with calories values > 500 and pulse < 100. 
print("rows with calories greater than 500 and pulse less tha 100\n",df[(df.Calories>500) & (df.Pulse<100)])
#Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”
df_modified=pd.DataFrame(df,columns=["Duration","Pulse","Calories"])
print("new dataframe that contain all columns except Maxpulse\n",df_modified)
#Delete the “Maxpulse” column from the main df dataframe
del df['Maxpulse']
print("Data frame after Maxpulse deletion\n",df)
#. Convert the datatype of Calories column to int datatype
df["Calories"] = df["Calories"].astype("int") 
print("dataframe after converting calories column to int\n",df)
#create a scatter plot for the two columns (Duration and Calories).
print("Scatter plot for duration and Calories\n",df.plot.scatter(x = 'Duration', y = 'Calories'))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
#Data to plot
Programminglanguages = 'Java','Python','PHP','JavaScript','C#','C++'
Popularity = [22.2,17.6,8.8,8,7.7,6.7]
colors = ["#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd","#8c564b"]
#explode the 1st wedge
explode = [0.1, 0, 0, 0, 0,0]
#plot the pie chart
plt.pie(Popularity,labels=Programminglanguages,explode=explode,colors=colors,shadow=True,startangle=140,
        wedgeprops={'edgecolor': 'black'},autopct='%1.1f%%')
#Give the title
plt.title('Popularity of Programming languages')
#Adjust the padding between and around subplots.
plt.tight_layout()
#display the pie chart
plt.show()
```
